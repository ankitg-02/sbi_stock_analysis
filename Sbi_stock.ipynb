```python
sbi_t=sbi.groupby("Time")
sbi_time=sbi_t.get_group("09:20:00")
```

```python
sbi_sorted_last=sbi_time.tail().sort_values(by="Volume",ascending=False)
print(sbi_sorted_last)
```

```python
sbi_sorted_last["Rank"] = [x for x in range(1,6)]
print(sbi_sorted_last)
```
